```python
import pandas as pd
import numpy as np
```

```python
vine_df=pd.read_csv("vine.csv")
vine_df
```

```python
total_votes_df=vine_df[vine_df["total_votes"]>=20]
total_votes_df
```

```python
helpful_df=total_votes_df[(total_votes_df["helpful_votes"]/total_votes_df["total_votes"])>=0.5]
helpful_df
```

```python
paid_df=helpful_df[helpful_df["vine"]=='Y']
paid_df
```

```python
nonpaid_df=helpful_df[helpful_df["vine"]=='N']
nonpaid_df
```

```python
total_reviews=helpful_df["total_votes"].count()
total_reviews
```

```python
paid_reviews=paid_df["total_votes"].count()
paid_reviews
```

```python
nonpaid_reviews=nonpaid_df["total_votes"].count()
nonpaid_reviews
```

```python
fivestar_all=helpful_df[(helpful_df["star_rating"]==5)].count()["total_votes"]
fivestar_all
```

```python
fivestar_paid=paid_df[(paid_df["star_rating"]==5)].count()["total_votes"]
fivestar_paid
```

```python
fivestar_nonpaid=nonpaid_df[(nonpaid_df["star_rating"]==5)].count()["total_votes"]
fivestar_nonpaid
```

```python
percent_all=(fivestar_all/total_reviews * 100)
percent_all
```

```python
percent_paid=(fivestar_paid/paid_reviews * 100)
percent_paid
```

```python
percent_nonpaid=(fivestar_nonpaid/nonpaid_reviews * 100)
percent_nonpaid
```

```python
summary_df = pd.DataFrame({
    "Total Reviews" : total_reviews,
    "Paid Reviews" : paid_reviews,
    "Non Paid Reviews" : nonpaid_reviews,
    "5-Star Total Reviews" : fivestar_all,
    "5-Star Paid Reviews" : fivestar_paid,
    "5-Star Non Paid Reviews" : fivestar_nonpaid,
    "% 5-Star Total" : percent_all,
    "% 5-Star Paid" : percent_paid,
    "% 5-Star Non Paid" : percent_nonpaid}, index=[0])

summary_df
```
